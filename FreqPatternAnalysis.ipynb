### FP-growth mined frequent patterns on the dyre log set 

```python
# frequent patterns from negative sequences and positive sequences (Dyre_sept)
posfn="data/positive_patterns.txt"
negfn="data/negative_patterns.txt"
posset = open(posfn,'r').readlines()[2:]
negset = open(negfn,'r').readlines()[2:]
posset[1:10]

# --------------------------
# Explanation of the output: 
# - each line is a frequent pattern set. 
# - The number at the end is the frequencies of the item set. 
# ---------------------------
```

```python
def parseFP(line):
        return line.split('],')[0].strip('[')
```

# Jaccard Similairty

```python
#Jaccard similarity gives the most basic metric between two itemsets. It is the ratio of Intersection over Union
```

```python
def jaccard(s1, s2):
        "takes two lists as input and returns Jaccard coefficient"
        st1=set(s1)
        st2=set(s2)
        u = set(st1).union(st2)
        i = set(st1).intersection(st2)
        return float(len(i))/len(u)
```

```python
# similarity between postive patterns and the negative patterns (sum of 1-to-many)
jac_vec={}
for posline in posset:
    p = filter(None, parseFP(posline))
    jac_vec[p]=[]
    for negline in negset:
        n = negline.split('],')[0].strip('[').split(',')
        jac = jaccard(p.split(','), n)
        jac_vec[p].append(jac)
        #print p,str(jac_vec[p])[1:-1]
sum_jac = {k: sum(v) for k,v in jac_vec.items()}
for k in sorted(sum_jac, key=sum_jac.get, reverse = False):
    print k, sum_jac[k]
    
# ----------------------
# Explanation of output: 
# - each row is a frequent itemset that came out of dyre user activities. 
# - The float at the end of the line is the similarity metric between the positive frequent itemset with the negative patterns
# - the ones with 0.0 as similarity are the ones that show up in dyre actitivies only. 
# ----------------------
```

## Similarity Matrix

```python
def getSimMatrix():
    jac_vec=[]
    ids = []

    for posline in posset:
        p = filter(None, parseFP(posline))
        jac_vec.append(['1'])
        ids.append(p) 

    for negline in negset:
        n = filter(None, parseFP(negline))
        jac_vec.append(['0'])
        ids.append(n)

    for j in range(len(ids)):
        row = jac_vec[j]
        for k in range(len(ids)):
                jac = jaccard(ids[j].split(','), ids[k].split(','))
                row.append(jac)
        #print str(jac_vec[j])[1:-1]
        jac_vec[j]=row
    return jac_vec
```

```python
print "number of positive cases",len(posset), "; number of negative cases", len(negset)
```

```python
from pylab import pcolor, show, colorbar, xticks, yticks
import numpy as np
import pandas as pd

simMat = getSimMatrix()
data=[simMat[i][2:] for i in range(len(ids))]
#data = np.array(data)
df = pd.DataFrame(data)
df
```
