# Analysis with Labeled Datasets

### FP-growth mined frequent patterns on the dyre log set 

```python
# frequent patterns from negative sequences and positive sequences (Dyre_sept)
posfn="data/positive_patterns.txt"
negfn="data/negative_patterns.txt"
posset = open(posfn,'r').readlines()[2:]
negset = open(negfn,'r').readlines()[2:]
posset[1:10]
```

```python
def parseFP(line):
        return line.split('],')[0].strip('[')
```

# Jaccard Similairty

```python
def jaccard(s1, s2):
        "takes two lists as input and returns Jaccard coefficient"
        st1=set(s1)
        st2=set(s2)
        u = set(st1).union(st2)
        i = set(st1).intersection(st2)
        return float(len(i))/len(u)
```

```python
#the sequences 
jac_vec={}
for posline in posset:
    p = filter(None, parseFP(posline))
    jac_vec[p]=[]
    for negline in negset:
        n = negline.split('],')[0].strip('[').split(',')
        jac = jaccard(p.split(','), n)
        jac_vec[p].append(jac)
        #print p,str(jac_vec[p])[1:-1]
sum_jac = {k: sum(v) for k,v in jac_vec.items()}
for k in sorted(sum_jac, key=sum_jac.get, reverse = False):
    print k, sum_jac[k]
```
